# Creating Charts (Object-Oriented Interface Method)

```python
# Object-Oriented Interface Method for creating charts

%matplotlib inline

import matplotlib.pyplot as plt

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
# This is a breakdown of the code in previous cell^ - output is the same

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis) 
```

# Annotate Charts (MATLAB Method)

```python
# Annotate charts with MATLAB method

# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Annotate charts with MATLAB method (cont.)

# Create the plot.
# changes width and color of line, adds markers for data points
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# MATLAB method SkillDrill

%matplotlib inline
import matplotlib.pyplot as plt
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis, y_axis, label = "Boston", marker = "D", color = "green", linewidth = 2)

plt.xlabel("Months")
plt.ylabel("Fares")
plt.title("Fares by Month")

plt.ylim(0,50)

plt.grid()

plt.legend()
```

# Scatter Plots and Bubble Charts:
# MATLAB Method

```python
# Scatter Plots (MATLAB)

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# can use plt.plot() or plt.scatter()
# need 'o' as a parameter inside plt.plot()

plt.plot(x_axis, y_axis, 'o')

plt.scatter(x_axis, y_axis)
```

```python
# Format scatter plot (MATLAB)

# switch x and y axis, add legend for Chicago, change markers to red
plt.scatter(y_axis, x_axis, label = "Chicago", color = "red", )

# add legend
plt.legend()

# add labels - NOTE: xlabel and ylabel are different from previous (since x and y axis were switched)
plt.title("Fare per Month")
plt.xlabel("Fares($)")
plt.ylabel("Months")

# add range limit for fares
plt.xlim(0,50)

# invert months so Jan is at top
plt.gca().invert_yaxis()
```

# Bubble Chart (MATLAB)

```python
# create bubble chart (MATLAB)
# use scatter plot but add the size attribute "s"

plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
# adjust size of markers by using for loop on y_axis data

# declare y_axis_larger as a list
y_axis_larger = []

# for each value in y_axis, multiply it by 3
for data in y_axis:
    y_axis_larger.append(data * 3)
    
# dislpay the adjusted plot
plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
# use "list comprehension" technique for changing marker size
# looks almost like a for loop
# s = [expression for item in list]
 
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

# Scatter Plot (OO Method)

```python
# scatter plot OO Method

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

# Bubble Chart (OO Method)

```python
# add "s" parameter to scatter plot

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = y_axis)
```

```python
# SkillDrill: format bubble chart using OO Method

fig, ax = plt.subplots()

# add legend label, change marker color to blue, change size of markers
# make marker color transparent, add black edge to markers
# switch x and y axis

ax.scatter(y_axis, x_axis, label = "Boston", s = [i * 5 for i in y_axis])

ax.legend()

ax.set_title("Fare by Month")
ax.set_xlabel("Fares")
ax.set_ylabel("Months")

ax.invert_yaxis()
```

# Pie Charts (MATLAB)

```python
# plot a pie chart - similar to above methods, but plt.pie() only takes one list of values,
# use y_axis list as values and x_axis list as labels
plt.subplots(figsize=(8,8))
plt.pie(y_axis, labels = x_axis)

# "explode" a wedge out of the pie chart
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# add percentages to chart
plt.pie(y_axis, explode = explode_values, labels=x_axis, autopct='%.1f%%')

# increase size of pie chart

```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

# Pie Charts (OO Method)

```python
# create pie chart 

fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
```

```python
# SkillDrill: Pie Charts OO Method

fig, ax = plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", 
          "yellowgreen", "greenyellow", "yellow", "orange", 
          "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       shadow=True,
       startangle=90,
       counterclock=False,
       labels=x_axis,
       autopct='%.1f%%')

plt.show()
```

# Chart Exras: Error Bars, Major Ticks, Minor Ticks

## Error Bars on Line Chart (MATLAB):

```python
# Error Bars (MATLAB)
# plt.errorbar(x,y, yerr=<>)value

# Start again, and import statistics dependency too
%matplotlib inline

# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# plot error bars showing standard deviation

plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
# format error bars to have caps

plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

## Error Bars on Line Chart (OO Method):

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

## Error Bars on Bar Chart (MATLAB):

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

## Error Bars on Bar Chart (OO Method):

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

## Major Ticks (MATLAB):

```python
# add major ticks to horizontal bar chart

# import numpy - it's a numerical mathematics library that can create arrays/matrices of numbers
import numpy as np

# create the bar graph
plt.barh(x_axis, y_axis)

# use numpy to create the array for ticks (every $5)
# use arrange() function to define the upper/lower limits and increment frequency
plt.xticks(np.arange(0, 51, step=5.0))

# invert the y axis to have Jan on top
plt.gca().invert_yaxis()
```

## Major Ticks (OO Method)

```python
# create the horizontal bar chart
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)

# make sure numpy is imported
# define array for increments of major ticks
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

## Minor Ticks (OO Method):

```python
# need to import MultipleLocator

# define increments for minor ticks

# x.xaxis.set_minor_locator()

from matplotlib.ticker import MultipleLocator

# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
